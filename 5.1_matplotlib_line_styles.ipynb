```python
# Line styles can be changed with the "linestyle","linewidth" and "color" arguments
# common line styles are solid,dashed and dotted
import matplotlib.pyplot as plt
import pandas as pd
from PIL.ImageColor import colormap

data=pd.read_csv('cereal.csv')
print(data.head())
# Plotting the Chart
fig, ax = plt.subplots()
ax.plot(data.index ,data['carbo'],linestyle='--',linewidth=1,color='blue')
ax.plot(data.index ,data['rating'],linestyle='dotted',linewidth=2,color='black')# Dotted also means ':'
ax.plot(data.index ,data['calories'],linestyle='-',linewidth=3,color='yellow')

# Formatting the Chart
fig.suptitle('Macro Nutrition',fontsize=20)
ax.set_title('Nutrition',fontsize=10)
ax.set_xlabel('Index',fontsize=10)
ax.set_ylabel('Carbohydrates,Calories and Rating',fontsize=10)
ax.legend(['carbo','rating','calories'])
plt.show()

```
