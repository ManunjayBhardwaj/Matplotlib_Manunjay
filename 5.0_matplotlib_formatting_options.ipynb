```python
import matplotlib
# Matplotlib has these formatting options by PyPlot and Object-Oriented plots

# The set_title() and ste_label() methods let you add chart titles and axis labels
# fig.suptitle() serves as an overall figure title
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('housing.csv')
data1=pd.DataFrame(data)
# print(data1)
fig, ax = plt.subplots()
ax.plot(data.index,data["longitude"])# this data.index specifies the value of the x-axis, while the next specifies the value of the y-axis
ax.plot(data.index,data["latitude"])

fig.suptitle("How the value of longitude and latitude vary")# this prints the label on the empty canvas
ax.set_title("Variation")# this gives the name to the box of the chart
ax.set_ylabel("Longitude And Latitude")# gives y label
ax.set_xlabel("Index")# gives x label

# We can modify chart font sizes with the "fontsize" argument : You can specify the size in points (10,11,12) or relative sizes are also there("smaller","x-large")
fig.suptitle("How the value of longitude and latitude vary",fontsize=15)# this prints the label on the empty canvas
ax.set_title("Variation",fontsize=10)# this gives the name to the box of the chart
ax.set_ylabel("Longitude And Latitude",fontsize=10)# gives y label
ax.set_xlabel("Index",fontsize=10)# gives x label
```

```python
# We can also give chart legends to each series by the legend() method
# the series labels are used by default, but custom values can also be passed through
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('housing.csv')
data2=pd.DataFrame(data)
# print(data2)
fig, ax = plt.subplots()
ax.plot(data.index,data["longitude"])# this data.index specifies the value of the x-axis, while the next specifies the value of the y-axis
ax.plot(data.index,data["latitude"])

fig.suptitle("How the value of longitude and latitude vary")# this prints the label on the empty canvas
ax.set_title("Variation")# this gives the name to the box of the chart
ax.set_ylabel("Longitude And Latitude")# gives y label
ax.set_xlabel("Index")# gives x label

# now we will give legend to the chart
ax.legend(["longitude","latitude"])

# we can change the location with the "loc" or "bbox_to_anchor" arguments
# loc lets you set a pre-determined location option
ax.legend(["longitude","latitude"],loc="center right")
# if you want to study the types of attributes that you can pass in the loc function refer to the pdf

# "bbox_to_anchor" lets you set specific (x,y) coordinates
ax.legend(["longitude","latitude"],bbox_to_anchor=(1,1))
# the (0,0) is the origin which is the bottom left corner to specify the position of the legend you can give any value from 0 to 1 in the two parameters
# if we give values greater than 1 then we can easily shift the legend outside our chart area( useful when there is no whitespace)

# if we want the legend to be horizontal then we can pass the ncol= attribute and if we want that our legend should not have any box on top of it then we should use frameon=False

ax.legend(["longitude","latitude"],bbox_to_anchor=(1,-0.1),ncol=2,frameon=False)
```
