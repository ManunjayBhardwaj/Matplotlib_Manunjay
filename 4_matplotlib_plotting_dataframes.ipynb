```python
# When plotting DataFrames using the Object-Oriented interface,Matplotlib will use the index as the x axis and plot each column as separate series by default
# Plotting each series independently allows for improved customization
# Syntax: ax.plot(x-axis series, y-series values)
# here if we pass the x axis series with the y value then it will plot the y-series values using the x-axis series as index

import pandas as pd
import matplotlib.pyplot as plt
data_dict={
    "Sales":[10,20,30,40],
    "Profit":[2,3,4,1],
    "Date":["2022-01-02","2022-01-03","2022-01-04","2022-01-05"]
}
# it is a pro tip using date as index for plotting
df=pd.DataFrame(data_dict,columns=["Sales","Profit"],index=data_dict["Date"])
print(df)

fig, ax = plt.subplots()
ax.plot(df,ls="--")# this ls"--" gives the output as dashed lines
ax.set_title("Profit vs Sales");
```
