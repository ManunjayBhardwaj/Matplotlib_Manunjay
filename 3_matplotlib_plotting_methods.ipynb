```python
# Matplotlib has two plotting methods, or interfaces
# PyPlot API and Object Oriented

#PyPlot API: Charts are created with the plot() function and modified with additional functions

import matplotlib.pyplot as plt
y=[x**2 for x in range(1,10)]
plt.plot(y)
```

```python
# Object Oriented Plotting : They are built by adding axes, or charts , to a figure
# The subplots() function lets you create the figure and axes in a single line of code and you can then use figure and axis methods to customize the different elements in the plot

import pandas as pd
data=[x*10 for x in range(0,10)]
labels=[f'{x}!' for x in range(0,10)]

# By this we do the same work as above but this approach improves customization
# fig=plt.figure()
# ax=fig.add_subplot()
# ax.plot(data)
# ax.set_title('Plot Title')

# we can also use
fig,ax=plt.subplots()
ax.plot(data)
ax.set_title('Plot Title');
# The semi colon here destructs the useless line above the plot
# figure will be the overall canvas while ax would be the  specific chart on that canvas
```
