```python
# Figure size: you can adjust the figure size with the "figsize" argument
# Syntax: figsize(width,height)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.core.pylabtools import figsize
from PIL.ImageColor import colormap

data=pd.read_csv('cereal.csv')
print(data.head())
# Plotting the Chart and Giving the figure size as an argument to the figplots function
fig, ax = plt.subplots(figsize=(10,10))# here
# this figszie increases the whitespace and helps in better formatting

ax.plot(data.index ,data['carbo'],linestyle='--',linewidth=1,color='blue')
ax.plot(data.index ,data['rating'],linestyle='dotted',linewidth=2,color='black')# Dotted also means ':'
ax.plot(data.index ,data['calories'],linestyle='-',linewidth=3,color='yellow')

# Formatting the Chart
fig.suptitle('Macro Nutrition',fontsize=20,y=0.95)
# this y=0.95 brings back the suptitile back to its original space even after the figsize is increased
ax.set_title('Nutrition',fontsize=10)
ax.set_xlabel('Index',fontsize=10)
ax.set_ylabel('Carbohydrates,Calories and Rating',fontsize=10)
ax.legend(['carbo','rating','calories'])
plt.show()

```

```python
# Custom X-ticks
# you can apply custom x-ticks with the set_xticks() and x_ticks() functions
# syntax: ax.set_ticks(iterable)
import matplotlib.pyplot as plt
import pandas as pd
import numpy as npy
from IPython.core.pylabtools import figsize
from PIL.ImageColor import colormap

data=pd.read_csv('cereal.csv')
print(data.head())
# Plotting the Chart and Giving the figure size as an argument to the figplots function
fig, ax = plt.subplots(figsize=(10,10))# here
# this figszie increases the whitespace and helps in better formatting

ax.plot(data.index ,data['carbo'],linestyle='--',linewidth=1,color='blue')
ax.plot(data.index ,data['rating'],linestyle='dotted',linewidth=2,color='black')# Dotted also means ':'
ax.plot(data.index ,data['calories'],linestyle='-',linewidth=3,color='yellow')

# Formatting the Chart
fig.suptitle('Macro Nutrition',fontsize=20,y=0.95)
# this y=0.95 brings back the suptitile back to its original space even after the figsize is increased
ax.set_title('Nutrition',fontsize=10)
ax.set_xlabel('Index',fontsize=10)
ax.set_ylabel('Carbohydrates,Calories and Rating',fontsize=10)
ax.legend(['carbo','rating','calories'])
x=npy.arange(start=0,stop=len(data.index),step=5)
ax.set_xticks(x)# by this we can specifically pass any list of values in the x axis for us to significantly analyse for example in the above example we analysed the x values in a table of 5
plt.xticks(rotation=45)# this function rotates the ticks of the x axis by any amount of degree
plt.show()



```
