```python
import matplotlib.pyplot as plt
import pandas as pd
data=[x*10 for x in range(0,10)]
print(data)
```

```python
labels=[f'{x}!' for x in range(0,10)]
print(labels)
```

```python
plt.plot(labels,data)
plt.show()
```

```python
df=pd.DataFrame(data, labels).assign(series2=lambda x:x[0]/2)
df.head()
plt.plot(df);
```
