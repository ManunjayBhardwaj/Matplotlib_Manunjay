```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

# Line charts are used for showing trends over time
# syntax:ax.plot(x-axis series,series values,formatting options)
# as done in the previous chapter


# Now we use stacked line charts which helps you visualize the overall trend over time
data=pd.read_csv("time_trends_10_columns.csv")
data['Date']=pd.to_datetime(data['Date'])
data['Value']=np.sort(data['Value'])
data['Stock']=np.sort(data['Stock'])
data['Expenses']=np.sort(data['Expenses'])
data['Date']=np.sort(data['Date'])

print(data.to_string())
fig, ax = plt.subplots()
# ax.plot(data.Date, data['Value'])
# ax.plot(data.Date, data['Expenses'])
# ax.plot(data.Date, data['Discount'])
# ax.plot(data.Date, data['Rating'])
# ax.plot(data.Date, data['Stock'])

# all this above work can be done by a simple stacked plot
ax.stackplot(data.Date, data['Value'], data['Stock'], data['Expenses'])
# ax.legend("Value","expenses","Stock")


```
