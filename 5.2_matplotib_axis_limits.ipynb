```python
# Axis limits: the set_xlim() and set_ylim() functions let you modify the axis limits
# Syntax: ax.set_xlim(lower limit,upper limit)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('cereal.csv')
print(data.head())
# Plotting the Chart
fig, ax = plt.subplots()
ax.plot(data.index ,data['carbo'],linestyle='-',linewidth=3,color='yellow')
ax.plot(data.index ,data['rating'],linestyle='dotted',linewidth=2,color='black')# Dotted also means ':'
# ax.plot(data.index ,data['calories'],linestyle='-',linewidth=3,color='yellow')
# Setting the x and y limits
ax.set_xlim(-1,70)
ax.set_ylim(-5,100)


# Formatting the Chart
fig.suptitle('Macro Nutrition',fontsize=20)
ax.set_title('Nutrition',fontsize=10)
ax.set_xlabel('Index',fontsize=10)
ax.set_ylabel('Carbohydrates And Calories',fontsize=10)
ax.legend(['carbo','rating','calories'])
plt.show()
```
